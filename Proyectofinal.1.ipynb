<h1 Style="color:#ce295e">Que tipo de propiedad tiene mayor facilidad de venta ?</h1>

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi=100)
 
make = gf.index.tolist()
y_pos = np.arange(len(make))
 
cant = gf['SalesRatio'].tolist()
 

barColors = ['#4b004c','#c70452','#e63c80', '#f0b9cf', '#e0d1ed', '#e0d1ed' ]
 

ax.barh(y_pos, cant, align='center', color = barColors)
ax.set_yticks(y_pos, labels=make)
ax.invert_yaxis() 
 
title = "Grafico 1: Que tipo de propiedad tiene mayor facilidad de venta ?"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 

stitle = "Â¿Cual es la propiedad con mayor porcentaje de venta?"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 11,
            fontweight = "regular",
            fontfamily = "serif")
 

etiqueta = "Cantidad casas vendidas"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "regular",
             fontfamily = "serif")
 

ax.xaxis.set_label_coords(x=0, y= -0.06, transform= ax.transAxes)
 

ax.tick_params(axis='both', 
               labelsize = 10)
 

plt.subplots_adjust(top=0.80)
 
 
plt.show()
```

<p Style="text-align: justify"> En base al grafico anteriormente representado el tipo de propierad con mayor  venta son  las propiedades residenciales con la mayoria de ventas 59,ooo+ de ventas. As su vez
la siquiente propiedad con mayor venta luedgo de las residenciales son las propiedades sin construcciones las cuales tiene un aproximado de 700 ventas</p>

<h1 Style="color:#ce295e">En cual ciudad o pueblo se dan mayor cantidad de ventas de propiedades? y que elementos particulares tienen esas ciudades o pueblos ? ?</h1>

```python
gf=ventvR.groupby(['Town']).count()
gf=gf.iloc[:,3:8]
gf=gf.sort_values('SalesRatio', ascending=False)
gf
```

```python
town_counts = ventvR['Town'].value_counts()

fig, ax = plt.subplots(1, 1, figsize=(19, 11), dpi=100)

sorted_town_counts = town_counts.sort_values(ascending=False)
highest_town = sorted_town_counts.index[0]

color_dict = {
    highest_town: '#c70452',  
}

base_color = '#f0b9cf' 

highlight_colors = [color_dict.get(town, base_color) for town in town_counts.index]
highlight_colors = [color if isinstance(color, str) else color[0] for color in highlight_colors]

town_counts.plot(kind='bar', color=highlight_colors, ax=ax)
ax.set_title('Number of Properties Sold in Each Town', fontsize=20, fontweight='bold', fontfamily='serif')

ax.set_xlabel('Nombre de la ciudad', fontsize=15, fontweight='regular', fontfamily='serif', labelpad=20)
ax.set_ylabel('Cantidad de Propiedades vendidas', fontsize=15, fontweight='regular', fontfamily='serif', labelpad=20)

ax.tick_params(axis='both', labelsize=10)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=10)

ax.xaxis.set_label_coords(x=0, y=-0.16, transform=ax.transAxes)

title = "Grafico 2: En cual ciudad o pueblo se dan mayor cantidad de ventas de propiedades?"
plt.suptitle(title,
             horizontalalignment='left',
             x=0.005,
             y=0.98,
             transform=fig.transFigure,
             fontsize=23,
             fontweight="bold",
             fontfamily="serif")

stitle = "Â¿Cual es el pueblo con mayor cantidad de ventas?"
ax.set_title(stitle,
              horizontalalignment='left',
              x=0.001,
              y=0.92,
              transform=fig.transFigure,
              fontsize=15,
              fontweight="regular",
              fontfamily="serif")

plt.tight_layout()
plt.show()


```
